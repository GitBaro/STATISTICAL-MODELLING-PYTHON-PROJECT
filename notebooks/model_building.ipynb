Build a regression model.

```python
# import numpy
import numpy as np
import pandas as pd
# import linear_model and datasets from sklearn
from sklearn import linear_model, datasets
from sklearn.datasets import fetch_california_housing
```

```python
pd.read_csv('yelp_merge.csv')
yelp_merge = pd.read_csv('yelp_merge.csv')
yelp_merge = yelp_merge.drop('Unnamed: 0', axis=1)
yelp_merge
```

```python
y = yelp_merge['rating']
x = yelp_merge[['searched_latitude', 'searched_longitude', 'review_count']]
```

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(x) # adding a constant
lin_reg = sm.OLS(y,X)
```

```python
model = lin_reg.fit()
```

Provide model output and an interpretation of the results. 

```python
print_model = model.summary()
print(print_model)
```

R-squared shows that the model explains a very small, only 1.7% of the variation in the results. 
p values shows that every independent variable is statistically significant as related to the rating, although very weak.
